## Tratando dados ausentes

* Importando bibliotecas

```python
import pandas as pd
import numpy as np
```

* Carregando o arquivo csv

```python
df = pd.read_csv('archive/covid_worldwide.csv')
```

* Visualizando o DF criado

```python
df.head()
```

* Verificando as informações do DF

```python
df.info()
```

* Salvando os indices das colunas em uma lista para conversão destas em categoria e float

```python
categorycal = df.columns[df.dtypes == object].to_list() 
df[categorycal[0]] = df[categorycal[0]].astype('category')
df.info()
```

```python
categorycal = df.columns[df.dtypes == object].to_list() 
df[categorycal] = df[categorycal].astype(str)

for i in range(len(categorycal)):
    df[categorycal[i]] = df[categorycal[i]].str.replace(',', '')

df[categorycal] = df[categorycal].astype(float)
df.head()
```

* Verificando a existência de dados nulos

```python
df.isna().sum()
```

* Substituindo os dados nulos pelas médias de suas respectivas colunas

```python
df = df.fillna(df.mean())
```

* Conferindo a tratativa dos dados nulos

```python
df.isna().sum()
```
