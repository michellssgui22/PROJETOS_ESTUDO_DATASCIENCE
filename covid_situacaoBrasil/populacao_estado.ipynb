> Para realização do calculo de mortalidade se faz necessário ter a população de cada estado.
Devido a tal, irei capturar de um site os dados faltantes.

```python
import pandas as pd
import numpy as np
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
html = urlopen('https://www.amunes.org.br/noticia/ler/173/ibge-divulga-nova-estimativa-populacional-dos-estados-e-municipios')
bs = BeautifulSoup(html.read(), 'html.parser')
populacao = bs.findAll('tr')
populacao[:5]
```

```python
tabela = []
for t in range(len(populacao)):
    tabela.append(populacao[t])
    
tabela[:5]
```

```python
df = pd.DataFrame(tabela)
df.drop(index=([0, 1]), inplace=True)
df = df.reset_index()
df.drop(columns=['index'], inplace=True)
```

```python
df
```

```python
for i in range(len(df[0])):
    df[0][i] = df[0][i].get_text()
    df[1][i] = df[1][i].get_text()
df
```

```python
df[1] = df[1].str.replace('.', '')
df =df.fillna(0)
df
```

```python
for i in range(len(df[0])):
    if df[1][i] == '' or df[1][i] == ' ':
        df[1][i] = 0
```

```python
df[1] = df[1].astype('int64')
df
```

```python
df = df[df[1] != 0]
df
```

```python
df = df.reset_index()
df.drop(columns=['index'], inplace=True)
```

```python
df = df.rename(columns={df.columns[0]:'estado'})
df = df.rename(columns={df.columns[1]:'populacao'})
```

```python
df = df.sort_values(by=['estado'])
df = df.reset_index()
df.drop(columns=['index'], inplace=True)
df
```

```python
df.to_csv('csv/populacao.csv')
```
