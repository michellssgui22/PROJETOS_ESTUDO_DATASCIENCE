```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from urllib.request import urlopen
import matplotlib.pyplot as plt
```

> Captura dos dados referente às vacinações realizadas em cada estado: https://especiais.g1.globo.com/bemestar/vacina/2021/mapa-brasil-vacina-covid/ 

> As vezes capturar dados direto do site se torna um pouco complexo e quando se trata de um baixo volume, podemos captura-lo direto do código fonte da página, exporta=lo para um arquivo e então carrega-lo no projeto.

```python
dados = pd.read_html('html_vacinacao.html')
df = pd.DataFrame(dados[0])
df
```

```python
df['2ª doseou doseúnica(Unidades)'] = df['2ª doseou doseúnica(Unidades)'].str.replace('.', '')
df['2ª doseou doseúnica(Unidades)'] = df['2ª doseou doseúnica(Unidades)'].astype('int64')
```

```python
df['2ª doseou doseúnica(Em %)'] = df['2ª doseou doseúnica(Em %)'].str.replace(',', '.')
df['2ª doseou doseúnica(Em %)'] = df['2ª doseou doseúnica(Em %)'].str.replace('%', '')
df['2ª doseou doseúnica(Em %)'] = df['2ª doseou doseúnica(Em %)'].astype(float)
```

```python
df['Dose dereforço*(Unidades)'] = df['Dose dereforço*(Unidades)'].str.replace('.', '')
df['Dose dereforço*(Unidades)'] = df['Dose dereforço*(Unidades)'].astype('int64')
```

```python
df['Dose dereforço*(Em %)'] = df['Dose dereforço*(Em %)'].str.replace(',', '.')
df['Dose dereforço*(Em %)'] = df['Dose dereforço*(Em %)'].str.replace('%', '')
df['Dose dereforço*(Em %)'] = df['Dose dereforço*(Em %)'].astype(float)
```

```python
df['1ª dose(Unidade)'] = df['1ª dose(Unidade)'].str.replace('.', '')
df['1ª dose(Unidade)'] = df['1ª dose(Unidade)'].astype('int64')
```

```python
df['1ª dose(Em %)'] = df['1ª dose(Em %)'].str.replace(',', '.')
df['1ª dose(Em %)'] = df['1ª dose(Em %)'].str.replace('%', '')
df['1ª dose(Em %)'] = df['1ª dose(Em %)'].astype(float)
```

```python
df['Total devacinasaplicadas'] = df['Total devacinasaplicadas'].str.replace('.', '')
df['Total devacinasaplicadas'] = df['Total devacinasaplicadas'].astype('int64')
```

```python
df = df.sort_values(by=['Total devacinasaplicadas'], ascending=False)
df
```

```python
df = df.reset_index()
df.drop(columns=['index'], inplace=True)
df
```

```python
df = df.rename(columns={'Estado': 'estado', 
                   '2ª doseou doseúnica(Unidades)': '2ª dose', 
                   '2ª doseou doseúnica(Unidades)': '2ª dose',
                   '2ª doseou doseúnica(Em %)': '2ª dose %',
                   'Dose dereforço*(Unidades)': 'reforco',
                   'Dose dereforço*(Em %)': 'reforco %',
                   '1ª dose(Unidade)': '1ª dose',
                   '1ª dose(Em %)': '1ª dose %',
                   'Total devacinasaplicadas': 'total vascinas'
                   }) 
                     
```

```python
df.to_csv('csv/vacinacao.csv')
```
