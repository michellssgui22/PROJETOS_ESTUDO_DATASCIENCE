```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

## Coleta de Dados
> ...

```python
html = urlopen('https://github.com/fivethirtyeight/data/blob/master/alcohol-consumption/drinks.csv')
bs = BeautifulSoup(html.read(), 'html.parser')
```

```python
dados = bs.findAll('tr', {'class': 'js-file-line'})
lista = []
n = 1
for i in dados:
    L = str('L' + str(n))
    new_dados = []
    new_dados.append(i.findAll('td'))
    lista.append(new_dados)
    n = n + 1
lista
```

```python
new_list = []
for i in range(len(lista)):
    new_list.append(lista[i][0])
new_list
```

```python
df = pd.DataFrame(new_list)
df
```

## Limpeza dos Dados
> ...

```python
df.drop(columns=[0], inplace=True)
df
```

```python
df = df.drop(0)
df
```

```python
df = df.rename(columns={1: 'pais', 2: 'porcoes_de_cerveja', 3: 'bebidas_espirituosas', 4: 'porcoes_de_vinho', 5: 'total_puro_alcool'})
df
```

```python
df = df.reset_index()
df.drop(columns=['index'], inplace=True)
df
```

```python
for i in range(len(df['pais'])):
    df['pais'][i] = df['pais'][i].get_text()
    df['porcoes_de_cerveja'][i] = df['porcoes_de_cerveja'][i].get_text()
    df['bebidas_espirituosas'][i] = df['bebidas_espirituosas'][i].get_text()
    df['porcoes_de_vinho'][i] = df['porcoes_de_vinho'][i].get_text()
    df['total_puro_alcool'][i] = df['total_puro_alcool'][i].get_text()

df
```

```python
df.dtypes
```

```python
df['porcoes_de_cerveja'] = df['porcoes_de_cerveja'].astype('int64')
df['bebidas_espirituosas'] = df['bebidas_espirituosas'].astype('int64')
df['porcoes_de_vinho'] = df['porcoes_de_vinho'].astype('int64')
df['total_puro_alcool'] = df['total_puro_alcool'].astype(float)
```
